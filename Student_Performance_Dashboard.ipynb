```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/students_score.csv')
df.head()
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df['NrSiblings'] = df['NrSiblings'].astype(int)
df.head()
```

```python
a = df.pivot_table(values=['MathScore','ReadingScore','WritingScore'],
                   index='EthnicGroup')
a
```

```python
b = pd.DataFrame(a)
```

```python
sns.heatmap(b,annot=True)
```
