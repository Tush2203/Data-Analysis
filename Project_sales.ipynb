```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('sales_data.csv',encoding='ISO-8859-1')
df.head()
```

# info()

```python
df.info()
```

# describe()

```python
df.describe()
```

# removing cols
---
- Status
- unnamed1
- age

```python
df.drop(['Status','unnamed1'],axis=1,inplace =True)
```

```python
df.drop(['Age'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Marital_Status']=df['Marital_Status'].map({1:'Unmarried',0:'Married'})
df.head()
```

```python
df['Gender']=df['Gender'].map({'F':'Female','M':'Male'})
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df
```

```python
df['Gender'].value_counts()
```

```python
a=sns.countplot(x='Gender',data=df)
for i in a.containers:
  a.bar_label(i)
```

```python
gp=df.groupby('Gender')['Amount'].sum()
gp
```

```python
df1=pd.DataFrame(gp)
a=sns.barplot(x='Gender',y='Amount',data=df1)
for i in a.containers:
  a.bar_label(i)
```

# For Age Group

```python
df['Age Group'].value_counts()
```

```python
ag = df.groupby('Age Group')['Amount'].sum()
ag
```

```python
df2 = pd.DataFrame(ag)
b=sns.barplot(x='Age Group',y='Amount',data=df2)
for i in b.containers:
  b.bar_label(i)
```

```python
plt.figure(figsize=(18,5))
c=sns.countplot(x='State',data=df)
for i in c.containers:
  c.bar_label(i)
```
